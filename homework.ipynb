```python
import pandas as pd
```

```python
df=pd.read_csv('kc-house-data (1) (1).csv')
```

```python
df.head()
```

```python
# Вычисляем среднюю цену за квадратный метр
avg_price_per_sqft = df["price"] / df["sqft_living"]

# Выводим среднюю цену за квадратный метр
print("Средняя цена за квадратный метр:", avg_price_per_sqft.mean())
```

```python
import seaborn as sns
import  matplotlib.pyplot as plt
```

```python
# Устанавливаем размер графика
plt.figure(figsize=(8, 6))

# Создаем распределение квадратуры жилой площади
sns.displot(df["sqft_living"])
# Устанавливаем заголовок и метки осей
plt.title("Распределение квадратуры жилой площади")
plt.ylabel("Частота")
plt.xlabel("Квадратура жилой площади")

# Отображаем гистограмму
plt.show()


```

```python
# Создаем график распределения года постройки
plt.figure(figsize=(10, 6))
sns.displot(df["yr_built"], kde=False)

# Называем график
plt.title("Распределение года постройки")

# Именуем оси x и y
plt.xlabel("Год постройки")
plt.ylabel("Количество домов")

# Выводим график
plt.show()
```

График показывает, что большинство домов в этом районе были построены в 1950-х и 1960-х годах. В 1970-е и 1980-е годы наблюдается небольшой рост объемов строительства, а затем в 1990-е годы объем строительства снова снижается. После 2000 года наблюдается небольшой рост строительства, но он не достигает уровня, наблюдавшегося в 1950-х и 1960-х годах.

```python
 #Создаем график распределения домов от наличия вида на набережную
plt.figure(figsize=(10, 6))
sns.countplot(x="view", data=df)

# Называем график
plt.title("Распределение домов от наличия вида на набережную")

# Именуем ось x
plt.xlabel("Наличие вида на набережную")

# Именуем ось y
plt.ylabel("Количество домов")

# Выводим график
plt.show()
```

В этом конкретном случае график показывает, что подавляющее большинство домов в этом районе не имеют вида на набережную. Только около 5% домов имеют такой вид.

```python
# Создаем гистограмму распределения этажей домов
plt.hist(df["floors"])

# Устанавливаем заголовок и метки осей
plt.title("Распределение этажей домов")
plt.xlabel("Этажи")
plt.ylabel("Количество домов")

# Отображаем гистограмму
plt.show()
```

```python
# Создаем гистограмму распределения состояния домов
plt.figure(figsize=(10, 6))
sns.countplot(x="condition", data=df)

# Устанавливаем заголовок и метки осей
plt.title("Распределение состояния домов")
plt.xlabel("Состояние")
plt.ylabel("Количество домов")

# Отображаем гистограмму
plt.show()
```

График показывает, что большинство домов в этом районе находятся в хорошем и очень хорошем состоянии. Очень небольшое количество домов находятся в плохом или аварийном состоянии.

```python
# Диаграмма 1: Распределение стоимости недвижимости
plt.figure(figsize=(10, 6))
sns.distplot(df["price"])
plt.title("Распределение стоимости недвижимости")
plt.xlabel("Стоимость")
plt.ylabel("Плотность")
plt.show()

# Диаграмма 2: Распределение квадратуры жилой площади
plt.figure(figsize=(10, 6))
sns.distplot(df["sqft_living"])
plt.title("Распределение квадратуры жилой площади")
plt.xlabel("Квадратура жилой площади")
plt.ylabel("Плотность")
plt.show()

# Диаграмма 3: Распределение года постройки
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x="yr_built", y="price")
plt.title("Распределение года постройки")
plt.xlabel("Год постройки")
plt.ylabel("Стоимость")
plt.show()

# Диаграмма 4: Распределение наличия вида на набережную
plt.figure(figsize=(10, 6))
sns.countplot(x="view", data=df)
plt.title("Распределение наличия вида на набережную")
plt.xlabel("Наличие вида на набережную")
plt.ylabel("Количество домов")
plt.show()

# Диаграмма 5: Распределение состояния домов
plt.figure(figsize=(10, 6))
sns.countplot(x="condition", data=df)
plt.title("Распределение состояния домов")
plt.xlabel("Состояние")
plt.ylabel("Количество домов")
plt.show()
```

Большинство домов в рассматриваемом районе имеют стоимость от 200 000 до 600 000 долларов США. Есть несколько домов, которые стоят более 1 миллиона долларов, но они являются выбросами

Большинство домов в этом районе имеют площадь от 1500 до 2500 квадратных футов. Есть несколько домов с площадью более 3000 квадратных футов, но они являются выбросами.

Большинство домов в этом районе были построены в 1950-х и 1960-х годах. В 1970-е и 1980-е годы наблюдается небольшой рост объемов строительства, а затем в 1990-е годы объем строительства снова снижается. После 2000 года наблюдается небольшой рост строительства, но он не достигает уровня, наблюдавшегося в 1950-х и 1960-х годах.

Подавляющее большинство домов в этом районе не имеют вида на набережную. Только около 5% домов имеют такой вид.

Большинство домов в этом районе находятся в хорошем или очень хорошем состоянии. Очень небольшое количество домов находятся в плохом или аварийном состоянии.

Эти диаграммы показывают, что рынок недвижимости в этом районе является относительно стабильным. Большинство домов имеют сопоставимую стоимость, площадь и возраст. Подавляющее большинство домов не имеют вида на набережную, но большинство домов находятся в хорошем состоянии.
